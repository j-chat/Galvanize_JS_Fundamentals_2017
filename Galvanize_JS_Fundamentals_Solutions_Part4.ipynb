## 38. JS some
Write a function named `some` that takes an array and a value and returns true if at least one element of the array equals the value
Example:
* If you pass `[1,2], 1` it should return `true`
* If you pass `[3,2], 1` it should return `false`

````javascript
function some(array, value) {
if (array.indexOf(value) === -1) {
    return false;
} else {
    return true;
}
}
````

    - `arr` (array of numbers)

    - `arr` (array of numbers)
* Return the median of all of the numbers.

Documentation:

* https://developer.mozilla.org/en-US/docs/Web/JavaScript/Reference/Global_Objects/Math/min
* https://developer.mozilla.org/en-US/docs/Web/JavaScript/Reference/Global_Objects/Array/sort

````javascript
function sum(arr) {
  var sum = 0;
  for (var i = 0; i < arr.length; i++) {
   sum += arr[i];
  }
  return sum;
}

function product(arr) {
    var prod = 1;
    for (var i = 0; i < arr.length; i++) {
        prod *= arr[i];
  }
  return prod;
}

function filterPassingGrades(arr) {
  var grades = [];
   for (let i = 0; i < arr.length; i++) {
     if (arr[i] > 69) {
       grades.push(arr[i]);
     }
   }
   return grades;
}

function replace(arr, from, to) {
  for (var i=0; i < arr.length; i++)  {
      if (arr[i] === from) {
          arr[i] = to;
      } 
  }
  return arr;
}

 function max(arr) {
  var max = arr[0];
  if (arr.length === 0) {
  return -Infinity;
} else {
  for (let i = 0; i < arr.length; i++) {
    if (max <= arr[i+1]) {
      max = arr[i+1];
    } 
  }
}
return max;
}

 function min(arr) {
  var min = arr[0];
  if (arr.length === 0) {
  return Infinity;
} else {
  for (let i = 0; i < arr.length; i++) {
    if (arr[i+1] <= min) {
      min = arr[i+1];
    } 
  }
}
return min;
} 

function mean(arr) {
  var sum = 0;
  if (arr.length === 0) {
    return null;
  } else {
  for (var i = 0; i < arr.length; i++) {
   sum += arr[i];
  }
  }
  return (sum/arr.length);
}

function median(arr) {
var arr = arr.sort();
  if (arr.length === 0) {
    return null;
} else if (arr.length % 2 == 1) {
  return arr[(arr.length - 1)/2];
} else {
  var a = arr[(arr.length/2)-1];
  var b = arr[(arr.length)/2];
  return (a+b)/2;
}
}
````

## 43. Iterate Over an Array Exercise 2
Adele is having trouble remembering her own song lyrics. For some reason, all she knows is the word "Hello". Let's help her out.

```var lines = [
    "It's me.",
    'Can you hear me?',
    'from the other side',
    'from the outside'
];```

Write a loop that logs to the screen each of her lines with the word "Hello" in front of it.

````javascript
var lines = [
    "It's me.",
    'Can you hear me?',
    'from the other side',
    'from the outside'
];

for (var i = 0; i < lines.length; i++) {
    var one_line = lines[i];
    console.log("Hello " + one_line);
}
````

## 44. Traverse an Object Exercise 1
Create an object called `person` and give it the values `{'name':'Bill', 'age': 20}`.
Write a loop that traverses the object and prints the string "name: Bill, age: 20".

````javascript
var person = {'name':'Bill', 'age': 20};

for (var key in person) {
  console.log(key + ": " + person[key]);
}
````
