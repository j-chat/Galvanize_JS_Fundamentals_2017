## 64. Identify Properly Structured CSS

### Improving Code Readability
The three samples of code below all contain the exact same code. But due to the different formatting of the code, which of these CSS examples will <b>not</b> work?

#### Sample 1:
 h1 {
  color: #1B676B;
  font-size: 32px;
  font-family: times;
}

p {
  color: #EAFDE6;
  font-size: 16px;
  font-family: arial;
}

#### Sample 2:
h1 {color:#1B676B;font size:32px;font family:times;}

p {color:#EAFDE6;font size:16px;font family:arial;}

#### Sample 3:
h1         { color:
#1B676B;        font-size:  
            32px;
font-family:    
                    times;      }

p
{     color: #EAFDE6;
font-size: 16px;
font-family:arial;          }

Answer: Despite the crazy formatting, the only problem that matters is the spaces in the property names in Sample 2. Remember: Spaces in code do not affect the code itself, unless they're found inside of a name. Although the code in Sample 3 works, that the code is  <b>not readable for other developers, or your instructors <\b>, so don't turn that kind of thing in!
